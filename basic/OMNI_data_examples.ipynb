# Load and plot OMNI data


```python
import pyspedas
import pytplot
from pytplot import tplot

pyspedas.omni.data( trange=['2017-03-24', '2017-03-31'] )
```

```python
pytplot.timespan( "2017-03-24 00:00:00", 7, keyword='days')
tplot(['SYM_H','AL_INDEX','BX_GSE', 'BY_GSM', 'BZ_GSM', 'flow_speed', 'proton_density', 'Pressure'])
```

```python
tplot( ['SYM_H','flow_speed', 'Pressure'], xsize=12, ysize=4 )
```

```python
pytplot.options('Pressure', 'ylog', True )
tplot( ['SYM_H','flow_speed', 'Pressure'], xsize=12, ysize=4 )
```

```python
pytplot.options('Pressure', 'ytitle', 'Dynamic¥npres.' )
tplot( ['SYM_H','flow_speed', 'Pressure'] )
```

```python
pytplot.options( 'AL_INDEX', 'yrange', [-1000,0])
tplot( ['SYM_H','AL_INDEX', 'Pressure'] )
pytplot.timebar(pyspedas.time_double('2017-03-26 20:00:00'))
```
